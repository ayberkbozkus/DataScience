```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify= cancer.target, random_state=66)
```

```python
train_truth = []
test_truth = []
```

```python
neighborhood_number=range(1,11)
```

```python
for n_neighborhood in neighborhood_number:
    class_ = KNeighborsClassifier(n_neighbors=n_neighborhood)
    class_.fit(X_train,y_train)
    train_truth.append(class_.score(X_train,y_train))
    test_truth.append(class_.score(X_test,y_test))
    
plt.plot(neighborhood_number, train_truth, label = 'Train Truth')
plt.plot(neighborhood_number, test_truth, label = 'Test Truth')
plt.ylabel('Truth')
plt.xlabel('n-neighborhood')
plt.legend()
```
